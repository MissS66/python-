## 五种方法创建dataframe

```python
#使用列表
import pandas as pd
data = [['1','宋江'],['2','卢俊义'],['3','吴用']]
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
df = pd.DataFrame(data,index=['rank1','rank2','rank3'],columns=['座次','姓名'])  #添加行索引和列索引
print(df)
```

```python
#使用字典创建dataframe
data = {'座次':['1','2','3'],'姓名':['宋江','卢俊义','吴用']}
df = pd.DataFrame(data,index=['rank1','rank2','rank3'])
print(df)
```

```python
#使用列表字典
data = [{'座次':'1','姓名':'宋江'},{'座次':'2','姓名':'卢俊义'},{'座次':'3','姓名':'吴用'}]
df = pd.DataFrame(data,index=['rank1','rank2','rank3'])
print(df)
```

```python
#使用series
s1 = pd.Series(['1','宋江'])
s2 = pd.Series(['2','卢俊义'])
s3 = pd.Series(['3','吴用'])
data = [s1,s2,s3]
df = pd.DataFrame(data,index=['rank1','rank2','rank3'])
print(df)
```

## DataFrame基本操作

### 操作列

```python

data = [{'座次':'1','姓名':'宋江'},{'座次':'2','姓名':'卢俊义'},{'座次':'3','姓名':'吴用'}]
df = pd.DataFrame(data,index=['rank1','rank2','rank3'])
print(df)
```

```python
# 新增列
df['绰号'] = ['及时雨','玉麒麟','智多星']
print(df)
```

```python
#修改列
df['绰号'] = ['呼保义','玉麒麟','智多星']
print(df)
```

```python
#获取一列
df['姓名']
```

```python
print(type(df['姓名']))
```

```python
#删除列
del df['绰号']
```

```python
df
```

```python
df.pop('绰号')
```

```python
df
```

```python
df.drop(columns=['绰号'])
```

```python
df
```

```python
#删除列总结
#del 直接删除源数据，没有返回值
#df.pop() 直接删除源数据，返回删除的series
#df.drop() 不删除源数据
```

### 操作行

```python
# 获取行
df.loc['rank1']  #返回的是转置后的series，为了方便操作。
```

```python
#iloc  通过整数值索引
rank2 = df.iloc[1]
print(rank2)
```

```python
#添加行
df.loc['rank4'] = ['4','公孙胜','入云龙']
print(df)
```

```python
df.loc['rank5'] = {'姓名':'关胜','绰号':'大刀','座次':'5'}
print(df)
```

```python
#修改行
df.loc['rank5'] = {'姓名':'关胜','绰号':'大刀关胜','座次':'5'}
print(df)
```

```python
#删除行
df1 = df.drop(index = ['rank3','rank5'])
print(df1)
print(df)
#drop方法会保留源数据
```

```python
df1 = df.drop(index = ['rank3','rank5'],inplace=True)  #不保留原始数据
print(df1)
print(df)
```

### 算术运算

```python
#Series
s1 = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
s2 = pd.Series([11,12,13,14,15],index=['a','b','c','d','e'])
print(s1+s2)
```

```python
s1 = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
s2 = pd.Series([11,12,13,14,15],index=['a','b','c','d','f'])
print(s1+s2)
```

```python
#DataFrame
import numpy as np
df1 = pd.DataFrame(np.arange(9).reshape(3,3),index=['宋江','李逵','武松'],columns=['语文','数学','英语'])
print(df1)
```

```python
df1['总成绩'] = df1['语文'] + df1['数学'] + df1['英语']
print(df1)
```

```python
df2 = pd.DataFrame(np.arange(9).reshape(3,3),index=['宋江','李逵','武松'],columns=['语文','数学','物理'])
print(df1+df2)
```

```python
#加法add，减法sub，乘法mul，除法 div
print(df1)
print(df2)
df = df1.add(df2,fill_value=0)  #为空的地方赋值为0
print(df)
```
