### 函数运用

```python
import numpy as np
import pandas as pd
df1 = pd.DataFrame(np.arange(9).reshape(3,3),index=['宋江','李逵','武松'],columns=['语文','数学','英语'])
print(df1)
```

```python
#使用numpy的通用函数
#获取列的最大数
np.max(df1)
#np.max(df1,axis=0)
```

```python
#获取行的最大数
np.max(df1,axis=1)
```

```python
#使用自定义函数
def text(df):
    return df.max()-df.min()

df1.apply(text)  
#pandas 的 apply() 函数可以作用于 Series 或者整个 DataFrame，
#功能是自动遍历整个 Series 或者 DataFrame, 对每一个元素运行指定的函数。
```

```python
def text(df):
    return df.max()-df.min()

df1.apply(text,axis=1)  
```

```python
df1.apply(lambda x:x.max() - x.min())  
```

```python
def total(df1):
    df1['总分'] = df1['语文'] + df1['数学'] + df1['英语']
    return df1
df1.apply(total,axis = 1)
```

```python
def total(df1):
    df1['总分'] = df1['语文'] + df1['数学'] + df1['英语']
    return df1
df1.apply(total,axis = 1)
```

```python
#添加自定义索引方法1
movies = pd.read_csv('data/豆瓣TOP250.txt',sep='|',header=None)
movies.columns=['电影名','评分','评论数']
movies
```

```python
#添加自定义索引方法2
movies = pd.read_csv('data/豆瓣TOP250.txt',sep='|',header=None,names =['电影名','评分','评论数'])
movies
```

```python
#文件存储为txt
movies.to_csv('豆瓣电影排名前250位.txt')
```

```python
#去除列索引及行索引
movies.to_csv('豆瓣电影排名前250位.txt',header=None,index=False)
```
